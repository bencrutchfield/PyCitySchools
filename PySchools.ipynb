```python
#Dependencies
import pandas as pd
import numpy as np
import os
```

```python
#define file path
schools_file = "raw_data/schools_complete.csv"
students_file = "raw_data/students_complete.csv"
```

```python
#Read CSV school file
schools_df = pd.read_csv(schools_file)
schools_df
```

```python
# Show just the header
schools_df.head()
```

```python
#Read CSV student file
file_two = "raw_data/students_complete.csv"
student_df = pd.read_csv(students_file)
```

```python
# Show just the header - Student
student_df.head()
```

```python
#Find the columns in the school csv
schools_df.columns
```

```python
#Find the columns in the student csv
student_df.columns
```

```python
#Rename school columns
schools_df.rename(columns = {'name': 'school'}, inplace = True)
schools_df.rename
```

```python
#Merge data in order for there not to be two "name" columns that display different types/information(data)
merged_df = student_df.merge(schools_df, how = 'left', on = 'school')
merged_df.head()
```

```python
#Count of Categoriges - School
different_school_name = schools_df["school"].unique()
different_school_name
```

```python
#Count of Categories - Student
student_category_df = student_df.count()
student_category_df.head()
```

```python
#Table for Student Categories
student_category_table = pd.DataFrame({"Student category": student_category_df,})
student_category_table.head()
```

```python
#Total Budget - District
total_budget = schools_df["budget"].sum()
total_budget
```

```python
#Average Math Score - District
math_score_average = student_df["math_score"].mean()
math_score_average
```

```python
#Average Reading Score - District
reading_score_average = student_df["reading_score"].mean()
reading_score_average
```

```python
#Percent Passing math
passing_math_score = student_df.loc[student_df["math_score"] >= 70, :]
passing_math_score.head()
```

```python
#Percent Passing reading
passing_reading_score = student_df.loc[student_df["reading_score"] >= 70, :]
passing_reading_score.head()
```

```python
#Total Students - District
total_student_number = student_df["name"].count()
total_student_number
```

```python
#Overall Passing Rate Percent
overall_pass_grade = student_df[(student_df['math_score'] >= 70) & (student_df['reading_score'] >= 70)]['name'].count()/total_student_number
overall_pass_grade
```

```python
#List of Different Schools - With categories
school_list= schools_df.groupby(["School ID"])
school_list.head()
```

```python
#List different schools by name
different_school_names = schools_df['school'].unique()
different_school_names
```

```python
#Gives unique data about schools
school_count = len(different_school_names)
school_count
```

```python
#Total district Dataframe 

district_summary = pd.DataFrame({
    
    "Total Schools": [school_count],
    "Total Students": [total_student_number],
    "Total Budget": [total_budget],
    "Average Reading Score": [reading_score_average],
    "Average Math Score": [math_score_average],
    "% Passing Reading":[passing_reading_score],
    "% Passing Math": [passing_math_score],
    "Overall Passing Rate": [overall_pass_grade]

})

district_summary
```

```python
#Create different df to change order
dist_sum = district_summary[["Total Schools", 
                             "Total Students", 
                             "Total Budget", 
                             "Average Reading Score", 
                             "Average Math Score", 
                             '% Passing Reading', 
                             '% Passing Math', 
                             'Overall Passing Rate']]

dist_sum
```

```python
dist_sum.style.format({"Total Budget": "${:,.2f}", 
                       "Average Reading Score": "{:.1f}", 
                       "Average Math Score": "{:.1f}", 
                       "% Passing Math": "{:.1%}", 
                       "% Passing Reading": "{:.1%}", 
                       "Overall Passing Rate": "{:.1%}"})
```

```python
#Sort by school
school_name = merged_df.set_index('school').groupby(['school'])
school_name.head()
```

```python
#School_types
school_types_name = schools_df.set_index('school')['type']
school_types_name.head()
```

```python
#total students by school
student_per_school = school_name['Student ID'].count()
student_per_school.head()
```

```python
#school budget
school_budget = schools_df.set_index('school')['budget']
school_budget.head()
```

```python
# budget per student 
student_budget = schools_df.set_index('school')['budget']/schools_df.set_index('school')['size']
student_budget.head()
```

```python
#Average math score
math_score_school = school_name['math_score'].mean()
math_score_school.head()
```

```python
#Average reading score
reading_score_school = school_name['reading_score'].mean()
reading_score_school.head()
```

```python
#Passing Math grade by school
pass_math_score = merged_df[merged_df['math_score'] >= 70].groupby('school')['Student ID'].count()/student_per_school
pass_math_score.head()
```

```python
#Passing Reading Grade by school
pass_reading_score = merged_df[merged_df['reading_score'] >= 70].groupby('school')['Student ID'].count()/student_per_school
pass_reading_score.head()
```

```python
#Passing Overall Score
overall_score = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['math_score'] >= 70)].groupby('school')['Student ID'].count()/student_per_school 
overall_score.head()
```

```python
school_summary = pd.DataFrame({
    "School Type": school_types_name,
    "Total Students": student_per_school,
    "Per Student Budget": student_budget,
    "Total School Budget": school_budget,
    "Average Math Score": math_score_school,
    "Average Reading Score": reading_score_school,
    '% Passing Math': pass_math_score,
    '% Passing Reading': pass_reading_score,
    "Overall Passing Rate": overall_score
})
```

```python
school_summary = school_summary[['School Type', 
                          'Total Students', 
                          'Total School Budget', 
                          'Per Student Budget', 
                          'Average Math Score', 
                          'Average Reading Score',
                          '% Passing Math',
                          '% Passing Reading',
                          'Overall Passing Rate']]
```

```python
sch_summary.style.format({'Total Students': '{:,}', 
                          "Total School Budget": "${:,}", 
                          "Per Student Budget": "${:.0f}",
                          'Average Math Score': "{:.1f}", 
                          'Average Reading Score': "{:.1f}", 
                          "% Passing Math": "{:.1%}", 
                          "% Passing Reading": "{:.1%}", 
                          "Overall Passing Rate": "{:.1%}"})
```
